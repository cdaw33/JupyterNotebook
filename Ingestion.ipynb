```python
#Imports everything needed for the system
from google.cloud import bigquery
import pandas as pd
import pydata_google_auth
credentials = pydata_google_auth.get_user_credentials(
    ['https://www.googleapis.com/auth/bigquery'],
)
import json
```

```python
#Sets the project ID then creates a BigQuery Client
project_id = "
client = bigquery.Client(project=project_id, credentials=credentials)
```

```python
#Creates a SQL statement that is then run using the BigQuery Client
sql = """
    SELECT *
    FROM `bigquery-public-data.chicago_crime.crime`
    WHERE Year = 2019    
"""
#Converts the data returned into a Pandas Dataframe
df2019 = client.query(sql).to_dataframe()
```

```python
sql = """
    SELECT *
    FROM `bigquery-public-data.chicago_crime.crime`
    WHERE primary_type = 'ASSAULT'    
"""

assaultdf = client.query(sql).to_dataframe()
```

```python
sql = """
    SELECT *
    FROM `bigquery-public-data.chicago_crime.crime`
    WHERE primary_type = 'THEFT'    
"""

theftdf = client.query(sql).to_dataframe()
```

```python
sql = """
    SELECT *
    FROM `bigquery-public-data.chicago_crime.crime`
    WHERE primary_type = 'NARCOTICS'    
"""

narcoticsdf = client.query(sql).to_dataframe()
```

```python
assaultcount = assaultdf["year"].value_counts().sort_index()
```

```python
finalassault = assaultcount.rename_axis('Year').reset_index(name='Count')
finalassaultjson = finalassault.to_json(orient='records')
```

```python
theftcount = theftdf["year"].value_counts().sort_index()
finaltheft = theftcount.rename_axis('Year').reset_index(name='Count')
finaltheftjson = finaltheft.to_json(orient='records')
```

```python
narcoticscount = narcoticsdf["year"].value_counts().sort_index()
finalnarcotics = narcoticscount.rename_axis('Year').reset_index(name='Count')
finalnarcoticsjson = finalnarcotics.to_json(orient='records')
```

```python
# GET /total_crime/assault
print(json.dumps(finalassaultjson))
```

```python
# GET /total_crime/theft
print(json.dumps(finaltheftjson))
```

```python
# GET /total_crime/narcotics
print(json.dumps(finalnarcoticsjson))
```

```python
sql = """
    SELECT District
    FROM `bigquery-public-data.chicago_crime.crime`
    WHERE year = 2018  
"""

districtdf = client.query(sql).to_dataframe()
```

```python
districtcount = districtdf["District"].value_counts().sort_index()
finaldistrict = districtcount.rename_axis('District').reset_index(name='Count')
finaldistrictjson = finaldistrict.to_json(orient='records')
```

```python
# GET /total_crime/district
print(json.dumps(finaldistrictjson))
```
